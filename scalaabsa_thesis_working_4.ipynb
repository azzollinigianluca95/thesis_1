```
%%writefile /content/Aspect-Based-Sentiment-Analysis/prova.py
from aspect_based_sentiment_analysis import aux_models
from aspect_based_sentiment_analysis import utils
from aspect_based_sentiment_analysis import loads
from aspect_based_sentiment_analysis import plots

text = ("the new boss was crazy: he went around the rooms in a crazy way. Employees were aware of that, and they always smile when they used to talk with him. Managers are good people always drink a coffee whit the boss seeming and agree with him, even if they didn't want. This is a perfect day: the boss is ok")
recognizer = aux_models.BasicPatternRecognizer()
nlp = loads.load(pattern_recognizer=recognizer)
completed_task = nlp(text=text, aspects=['boss', 'manager'])
subscriptions, price = completed_task.examples

plots.summary(price)
plots.display(price.review)
plots.summary(subscriptions)
plots.display(subscriptions.review)
```
